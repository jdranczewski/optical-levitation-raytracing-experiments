```python
with open("sphere.obj") as f:
    lines = f.readlines()
```

```python
verts = np.array([np.array(l[2:-1].split(" ")).astype(float) for l in lines if l[:2]=='v '])*.25
print(verts)
```

```python
faces = [[int(v.split("/")[0])-1 for v in l[2:-1].split(" ")] for l in lines if l[:2]=='f ']
print(len(faces))
```

```python
verts[faces[0]]
```

```python
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111, projection='3d')

for f in faces:
    t = pt.Triangle([0,0,0], *verts[f])
    t.plot(ax)

ax.plot(*verts.T, "x", c="tab:orange")
    
ax.set_xlim(-1.2,1.2)
ax.set_ylim(-1.2,1.2)
ax.set_zlim(-1.2,1.2)
```
